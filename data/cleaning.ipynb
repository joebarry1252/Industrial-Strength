```python
import pandas as pd
import os
```

```python
df = pd.DataFrame()
for file in os.listdir('rawdata'):
    fruit_type = file[:-4]
    tempdf = pd.read_csv(f"rawdata/{file}")
    
    tempdf["type"] = fruit_type
    tempdf = tempdf.drop('id', axis='columns')
    
    df = df.append(tempdf)
```

```python
df.to_csv("cleandata/fruits.csv")
```
